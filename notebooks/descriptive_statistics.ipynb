# Descriptive & Inferential Statistic

```python
# Import Libraries
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
# Specify the Plot style
plt.rcParams['figure.figsize'] = (10, 7.5)
```

```python
# Current working directory
cwd = os.getcwd()

# Parent directory
parent = os.path.dirname(cwd)

# Database directory
db = parent + '/database'

# Data file
shampoo_df = pd.read_csv(db + '/shampoo.csv')
```

## Inferential Statistics & Visualizations

```python
# Data Set Details
print (shampoo_df.info())
print (shampoo_df.describe().round(3))
```

```python
# Histograms for shampoo.df
plt.hist(shampoo_df['Sales'])
plt.title('Histogram of Shampoo Sales Data', fontsize=14)
plt.xlabel('Shampoo Sales($M)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
sns.distplot(shampoo_df['Sales'], bins = 20, hist=True, kde=True)
plt.title('Distribution of Shampoo Sales Data', fontsize=14)
plt.xlabel('Shampoo Sales($M)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()
```
